Analyze the data

```python
#Compare ratio between male and female

ax = sns.countplot(x='sex', data = df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

Find number of Smokers by Sex

```python
ax = sns.countplot(x='sex', data = df)
com_smoker = df.groupby(['sex', 'smoker'], as_index= False)['charges'].sum().sort_values(by= 'charges', ascending= False)
sns.barplot(x ='sex', y= 'charges', hue = 'smoker', data = com_smoker)

plt.show()
```

 Calculate which gender pay more medical charges 

```python

health_insurance_char = df.groupby(['sex'], as_index= False)['charges'].sum().sort_values(by= 'charges', ascending= False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x ='sex', y= 'charges', data = health_insurance_char)

plt.show()
```

 Distribution of medical charges across different regions

```python

region_insur = df.groupby(['region'], as_index= False)['charges'].sum().sort_values(by= 'charges', ascending= False)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x ='region', y= 'charges', data = region_insur)

plt.show()
```

compare smoker ratio in state

```python
ax = sns.countplot(data = df, x  = 'region', hue = 'smoker')

for bars in ax.containers:
  ax.bar_label(bars)

plt.show()
```

 Number of children affect medical charges

```python
children = df.groupby(['children'], as_index= False)['charges'].mean().sort_values(by= 'charges', ascending= False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x ='children', y= 'charges', data = children)

plt.show()
```

 Compared medical charges to smoker and non-smokers

```python
charges = df.groupby(['smoker'], as_index= False)['charges'].sum().sort_values(by = 'charges', ascending = False)

sns.barplot(x = 'smoker', y = 'charges',  hue = 'smoker', data = charges)

plt.show()
```

Average BMI for smokers vs. non-smokers

```python
charges = df.groupby(['smoker'], as_index= False)['bmi'].mean().sort_values(by = 'bmi', ascending = False)

sns.barplot(x = 'smoker', y = 'bmi',  hue = 'smoker', data = charges)

plt.show()
```
